```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
zomato_link="https://www.zomato.com/mumbai/ettarra-1-juhu/order"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}

r = requests.get(zomato_link, headers=headers)
```

```python
r.status_code
```

```python
soup=BeautifulSoup(r.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
results=soup.select('h4[class*="sc-1s0saks-15"]')
```

```python
results
```

```python
sc-17hyc2s-1 cCiQWA
```

```python
prices=soup.select('span[class*="sc-17hyc2s-1 cCiQWA"]')
```

```python
len(prices)==len(results)
```

```python
data={'prices':[],"items":[]}
```

```python
for i in range(len(prices)):
    data["prices"].append(prices[i].text)
    data['items'].append(results[i].text)
    
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df['prices']=df.prices.apply(lambda x: float(x))
```

```python
df
```

```python
df.to_csv("ettara_items.csv",index=False)
```
