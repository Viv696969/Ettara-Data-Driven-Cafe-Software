```python
import sqlalchemy
import pandas as pd
```

```python
engine=sqlalchemy.create_engine("mysql+mysqlconnector://:@:/")
```

```python
df=pd.read_sql_table("sentiment_analysis",engine)
```

```python
df
```

```python
df_monthly_review=pd.read_sql_query(
'''
select month(date) as month,round(avg(value),2) as avg_value
from sentiment_analysis
group by month(date)
order by month(date)
;
''',engine
)
```

```python
df_monthly_review.to_dict('records')
```

```python
df_yearly_review=pd.read_sql_query(
'''
select year(date) as year,round(avg(value),2) as avg_value
from sentiment_analysis
group by year(date)
order by year(date)
;
''',engine
)
```

```python
df_yearly_review.to_dict("list")
```

```python
sentiment_mapping={
    1:'very negative',
    2:'negative',
    3:'neutral',
    4:'positive',
    5:'excellent'
}
```

```python
df_yearly_review['average_sentiment']=df_yearly_review['avg_value'].apply(lambda x : sentiment_mapping[round(x)])
```

```python
df_yearly_review
```

```python
df_sentiment_count=pd.read_sql_query(
'''
select sentiment,count(sentiment) as `sentiment_count`
from sentiment_analysis
group by sentiment
order by `sentiment_count`;
 ''',engine)
```

```python

df_sentiment_count
```

```python
overall_avg_sentiment_count=pd.read_sql_query(
    'select round(avg(value)) as avg_sentiment_count from sentiment_analysis;',engine
)
```

```python
overall_avg_sentiment_count.to_dict('records')
```

```python
monthly_sentiment_count=pd.read_sql_query('''
select month(date) as `month`,sentiment,count(sentiment) as `monthly_count`
from sentiment_analysis group by `month`,`sentiment`
order by `month`;
''',engine)
```
